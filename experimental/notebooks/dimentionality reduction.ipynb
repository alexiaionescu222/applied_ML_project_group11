# Dimensionality reduction

Scale all datapoints first.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import os

print(os.getcwd())

data = pd.read_csv('../data/train_data.csv')
train = data[['normalized_lap','average_normalized_lap','lap_progress','current_position_norm']]

ground = data['finishing_position']

scaler = StandardScaler()
scaled = scaler.fit_transform(train)
```

Apply PCA

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(scaled)

# Visualize
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=ground, cmap='viridis', s=10)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.colorbar(label='Finishing Position')
plt.title('PCA of F1 Race Data')
plt.show()
```
